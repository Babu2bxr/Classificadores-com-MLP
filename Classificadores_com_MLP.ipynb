```python
!pip install -q seaborn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# 1. Carregamento e pré-processamento dos dados
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train.astype('float32') / 255.0
x_test  = x_test.astype('float32') / 255.0

# MLP espera vetor, não imagem 3D
x_train_flat = x_train.reshape((x_train.shape[0], -1))
x_test_flat = x_test.reshape((x_test.shape[0], -1))

# One-hot encoding dos rótulos para treinamento
y_train_cat = to_categorical(y_train, 10)
y_test_cat  = to_categorical(y_test, 10)

# 2. Função para construir modelos MLP variados
def build_mlp(units, hidden_layers, activation):
    model = Sequential()
    model.add(Dense(units, activation=activation, input_shape=(3072,)))
    for _ in range(hidden_layers - 1):
        model.add(Dense(units, activation=activation))
    model.add(Dense(10, activation='softmax'))
    return model

# 3. Definição dos 5 classificadores MLP
arquiteturas = [
    {'nome': 'MLP1', 'units': 128, 'hidden_layers': 1, 'activation': 'relu'},
    {'nome': 'MLP2', 'units': 256, 'hidden_layers': 2, 'activation': 'relu'},
    {'nome': 'MLP3', 'units': 128, 'hidden_layers': 3, 'activation': 'tanh'},
    {'nome': 'MLP4', 'units': 512, 'hidden_layers': 1, 'activation': 'sigmoid'},
    {'nome': 'MLP5', 'units': 64,  'hidden_layers': 2, 'activation': 'relu'}
]

modelos = []
for arq in arquiteturas:
    modelos.append(build_mlp(arq['units'], arq['hidden_layers'], arq['activation']))

# 4. Treinamento e avaliação
resultados = []
matrizes_confusao = []
metricas = []

for idx, modelo in enumerate(modelos):
    print(f"Treinando {arquiteturas[idx]['nome']} ...")
    modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    modelo.fit(x_train_flat, y_train_cat, epochs=10, batch_size=128, verbose=2, validation_split=0.1)

    # Previsão
    y_pred_probs = modelo.predict(x_test_flat)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = y_test.flatten()

    # Relatório de classificação
    relatorio = classification_report(y_true, y_pred, output_dict=True, zero_division=0)
    resultados.append(relatorio)

    # Métricas globais
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    rec = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')
    metricas.append([arquiteturas[idx]['nome'], acc, prec, rec, f1])

    # Matriz de confusão
    matriz = confusion_matrix(y_true, y_pred)
    matrizes_confusao.append(matriz)

    # Visualização da matriz de confusão
    plt.figure(figsize=(8,6))
    sns.heatmap(matriz, annot=False, cmap='Blues', fmt='d')
    plt.title(f'Matriz de Confusão - {arquiteturas[idx]["nome"]}')
    plt.xlabel('Predito')
    plt.ylabel('Verdadeiro')
    plt.show()

# 5. Tabela comparativa das métricas
import pandas as pd
tabela_metricas = pd.DataFrame(metricas, columns=['Modelo', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
print('\nTabela comparativa das métricas:')
display(tabela_metricas)

# 6. Exemplo de visualização das matrizes (opcional, já mostrado acima)
# Para mostrar todas as matrizes lado a lado:
plt.figure(figsize=(15,10))
for i, matriz in enumerate(matrizes_confusao):
    plt.subplot(2, 3, i+1)
    sns.heatmap(matriz, annot=False, cmap='Blues', fmt='d')
    plt.title(arquiteturas[i]['nome'])
    plt.xlabel('Predito')
    plt.ylabel('Verdadeiro')
plt.tight_layout()
plt.show()
```
